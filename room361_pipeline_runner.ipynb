# Room 361 Occupancy Pipeline Runner

This notebook is a thin runner for the modular Room 361 weak-supervision pipeline.
It calls the Python pipeline, loads the saved CSV artifacts, and displays the report-ready figures.

```python
from pathlib import Path

import pandas as pd
from IPython.display import Image, Markdown, display

from scripts.room361_pipeline import Room361PipelineConfig, run_room361_pipeline

OUTPUT_DIR = Path("reports/room361_pipeline")
result = run_room361_pipeline(Room361PipelineConfig(output_dir=str(OUTPUT_DIR)))

regression = pd.read_csv(OUTPUT_DIR / "model_comparison_regression.csv")
classification = pd.read_csv(OUTPUT_DIR / "model_comparison_classification.csv")
best_anchor = pd.read_csv(OUTPUT_DIR / "best_anchor_window_evaluation.csv")

display(Markdown("## Regression Comparison"))
display(regression)
display(Markdown("## Classification Comparison"))
display(classification)
display(Markdown("## Held-Out Anchor Evaluation"))
display(best_anchor)

print(f"Saved artifacts to {OUTPUT_DIR.resolve()}")
```

```python
FIGURE_DIR = OUTPUT_DIR / "figures"
figure_paths = [
    FIGURE_DIR / "room361_co2_flow_occupancy_timeseries.png",
    FIGURE_DIR / "room361_model_predictions_vs_anchors.png",
    FIGURE_DIR / "room361_anchor_residual_error.png",
    FIGURE_DIR / "room361_best_regression_feature_importance.png",
    FIGURE_DIR / "room361_engineered_feature_correlation_heatmap.png",
]

for figure_path in figure_paths:
    if not figure_path.exists():
        continue
    display(Markdown(f"## {figure_path.stem.replace('_', ' ').title()}"))
    display(Image(filename=str(figure_path)))
```
