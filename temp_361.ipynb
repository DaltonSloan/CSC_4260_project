```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data_361.csv')
df = df.dropna()

df['time'] = pd.to_datetime(df['time'])

df['time'] = df['time'].sort_values()

df.size
print(len(df))
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='time', y='Discharge Air Temp_value', data=df)
sns.lineplot(x='time', y='Zone Temp_value', data=df)
plt.title('Discharge Air and Zone Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (F)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
